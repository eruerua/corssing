### 下载肖申克的救赎照片

```python
import requests
url=' https://api.douban.com/v2/movie/1292052?apikey=0df993c66c0c636e29ecbb5344252a4a'
r = requests.get(url)
message=r.json()
print(message)
```

```python
pic_url=message['image']
print(pic_url)
```

```python
pic=requests.get(pic_url)
with open('Shawshank.jpg','wb') as file:
    file.write(pic.content)
```

### 下载豆瓣250清单

```python
import requests
top_250_list=[]
for i in range(0,250,20):
    url='  https://api.douban.com/v2/movie/top250?start='+str(i)+'&apikey=0df993c66c0c636e29ecbb5344252a4a'
    r = requests.get(url)
    message=r.json()
    #print(message['subjects'])
    [top_250_list.append([x['id'],x['title'],x['rating']['average'],','.join([x['name'] for x in x['casts']]),x['images']['large']]) for x in message['subjects']]
#print(top_250_list)
```

```python
import csv
print(top_250_list)
with open('top_250.csv','wt') as f:
    cw= csv.writer(f)
    for item in top_250_list:
        cw.writerow(item) #将列表的每个元素写到csv文件的一行
```
